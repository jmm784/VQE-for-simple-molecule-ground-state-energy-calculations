# Classical results for comparison:

```python
# Hartree-Fock classical results


from pyscf import gto, scf
import numpy as np


molecules = {
    "H": "H 0 0 0; H 0 0 0.74",          
    "LiH": "Li 0 0 0; H 0 0 1.6",         
    "BeH": "Be 0 0 0; H 0 0 1.3; H 0 0 -1.3"  
}

# Calculate HF energies
print("{:<6} {:<15} {:<10}".format("Molecule", "HF Energy (Ha)", "Converged?"))
print("="*40)

for name, geom in molecules.items():
    mol = gto.M(
        atom=geom,
        basis='sto-3g',
        unit='angstrom',
        verbose=0  
    )
    mf = scf.RHF(mol)
    mf.kernel()  # Run HF calculation
    
    # Format output
    energy = mf.e_tot
    converged = "" if mf.converged else ""
    print("{:<6} {:<15.6f} {:<10}".format(name, energy, converged))


print("\nNote: Calculations used STO-3G basis set and RHF method.")
```

```python
# Moller-Plesset classical results

from pyscf import gto, scf, mp

# Molecular geometries (in Angstroms)
molecules = {
    "H": "H 0 0 0; H 0 0 0.74",
    "LiH": "Li 0 0 0; H 0 0 1.6",
    "BeH": "Be 0 0 0; H 0 0 1.3; H 0 0 -1.3"
}

print("{:<6} {:<20} {:<10}".format("Molecule", "MP2 Energy (Ha)", "Converged?"))
print("="*45)

for name, geom in molecules.items():
    mol = gto.M(atom=geom, basis='sto-3g', unit='angstrom', verbose=0)
    
    # Run HF (required internally for MP2) then MP2
    mf = scf.RHF(mol).run()
    mp2_results = mp.MP2(mf).run()
    
    # Extract total MP2 energy (HF + correlation)
    mp2_energy = mf.e_tot + mp2_results.e_corr
    converged = "" if mf.converged else ""
    
    print("{:<6} {:<20.8f} {:<10}".format(name, mp2_energy, converged))

print("\nNote: STO-3G basis set | MP2 energy includes HF reference")
```
